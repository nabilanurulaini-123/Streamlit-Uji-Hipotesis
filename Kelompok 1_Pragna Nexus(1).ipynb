```python
%%writefile norangantuk.py
import streamlit as st
import numpy as np
import pandas as pd
from scipy import stats

st.sidebar.title("ğŸ“Š Menu Utama")

main_menu = st.sidebar.radio(
    "Navigasi:",
    [
        "ğŸ  Home",
        "ğŸ“„ Flowchart",
        "ğŸ“š Pilih Uji Statistik"
    ]
)

if main_menu == "ğŸ“š Pilih Uji Statistik":
    menu = st.sidebar.radio(
        "Pilih Jenis Uji Statistik:",
        [
            "Uji Proporsi 1 Sampel",
            "Uji Proporsi 2 Sampel",
            "Uji Rata-rata 1 Sampel (Z)",
            "Uji Rata-rata 1 Sampel (t)",
            "Uji Kesamaan Varians (F-test)",
            "Uji Rata-rata 2 Sampel Independen (Z-test)",
            "Uji Rata-rata 2 Sampel Independen (Pooled t-test)",
            "Uji Rata-rata 2 Sampel Independen (Welch t-test)",
            "Uji Rata-rata 2 Sampel Dependen (Paired t-test)"
        ]
    )
else:
    menu = None

st.title("ğŸ“˜ Dashboard Pengujian Hipotesis")
st.write("Aplikasi Proyek Akhir Pemrograman Komputer â€” Dashboard Uji Statistik")

if main_menu == "ğŸ  Home":
    st.header("Selamat Datang ğŸ‘‹")
    st.write("""
        Dashboard ini dibuat untuk memenuhi proyek akhir mata kuliah *Pemrograman Komputer*.
        
        Di dalam aplikasi ini, Anda dapat melakukan berbagai macam *uji statistika parametrik*, lengkap dengan:
        - Penjelasan konsep
        - Rumus matematis
        - Contoh perhitungan
        - Perhitungan otomatis (coming soon)
        
        Silakan pilih menu di sebelah kiri untuk mulai.
    """)

elif main_menu == "ğŸ“„ Flowchart":
    st.header("ğŸ“„ Flowchart Pengujian Hipotesis")
    st.write("Berikut flowchart yang diberikan pada soal:")
    
    # Pastikan file gambar ada, atau gunakan placeholder jika error
    try:
        st.image("flowchart.jpg", caption="Flowchart Pengujian Hipotesis", use_container_width=True)
    except:
        st.warning("Gambar flowchart.jpg tidak ditemukan di folder yang sama.")

def load_template(title):
    st.header(title)
    
    tab1, tab2, tab3, tab4 = st.tabs(
        ["Penjelasan", "Rumus", "Contoh", "Output Perhitungan"]
    )

    with tab1:
        st.subheader("ğŸ“˜ Penjelasan")
        st.info("Tuliskan penjelasan konsep uji statistik di sini.")

    with tab2:
        st.subheader("ğŸ“ Rumus Statistik Uji")
        st.warning("Masukkan rumus dari PDF ke sini (bisa latex).")

    with tab3:
        st.subheader("ğŸ§® Contoh Perhitungan")
        st.success("Berikan contoh manual sederhana.")

    with tab4:
        st.subheader("ğŸŸ¢ Perhitungan")
        st.write("Nantinya berisi input user + hasil perhitungan otomatis.")
        
# UJI PROPORSI 1 SAMPEL
# =========================
def load_proporsi_1_sampel(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    # ================= TAB 1 =================
    with tab1:
        st.subheader("Konsep Uji Proporsi 1 Sampel")
        st.write(r"""
        **Uji proporsi satu sampel** digunakan untuk menguji apakah **proporsi populasi ($p$)** 
        sama dengan nilai tertentu ($p_0$).
        
        Uji ini menggunakan **distribusi normal (Z)** sebagai pendekatan.
        """)

        st.info(r"""
        **Syarat Penggunaan:**
        1. Sampel diambil secara acak.
        2. Data bersifat dikotomi (suksesâ€“gagal).
        3. $np_0 \ge 5$ dan $n(1-p_0) \ge 5$.
        """)

    # ================= TAB 2 =================
    with tab2:
        st.subheader("Rumus Statistik Uji Proporsi 1 Sampel")

        st.latex(r"""
        Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}
        """)

        st.write("Keterangan:")
        st.markdown(r"""
        - $\hat{p} = \frac{x}{n}$ : proporsi sampel  
        - $x$ : jumlah sukses  
        - $n$ : jumlah sampel  
        - $p_0$ : proporsi hipotesis  
        """)

        st.subheader("Kriteria Pengujian (P-value)")
        st.write(r"""
        - **Tolak $H_0$** jika $P$-value $< \alpha$  
        - **Gagal Tolak $H_0$** jika $P$-value $\ge \alpha$  
        """)

    # ================= TAB 3 =================
    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Sebuah pabrik mengklaim bahwa **60% produknya cacat ringan**.
        Diambil sampel 100 produk, ternyata terdapat 55 yang cacat ringan.
        Uji apakah klaim pabrik benar pada taraf signifikansi 5%.
        """)

        st.markdown("**Perhitungan Manual:**")
        st.latex(r"""
        \hat{p} = \frac{55}{100} = 0.55
        """)
        st.latex(r"""
        Z = \frac{0.55 - 0.60}{\sqrt{\frac{0.6(1-0.6)}{100}}}
        = \frac{-0.05}{0.049} = -1.02
        """)

        st.write("P-value â‰ˆ 0.307 â†’ **Gagal Tolak H0**")

    # ================= TAB 4 =================
    with tab4:
        st.subheader("Kalkulator Uji Proporsi 1 Sampel")

        st.write("### 1. Masukkan Parameter")
        col1, col2 = st.columns(2)

        with col1:
            x = st.number_input("Jumlah Sukses (x)", min_value=0, step=1)
            n = st.number_input("Jumlah Sampel (n)", min_value=1, step=1)

        with col2:
            p0 = st.number_input("Proporsi Hipotesis (pâ‚€)", value=0.5, format="%.3f")
            alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])

        jenis_uji = st.selectbox(
            "Jenis Hipotesis Alternatif (H1)",
            ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"]
        )

        st.markdown("---")

        if st.button("Jalankan Uji Proporsi ğŸš€"):
            if x > n:
                st.error("Jumlah sukses tidak boleh lebih besar dari n!")
            else:
                # Hitung proporsi sampel
                p_hat = x / n

                # Hitung Z
                Z = (p_hat - p0) / np.sqrt((p0 * (1 - p0)) / n)

                # Hitung P-value
                if jenis_uji == "Dua Arah (â‰ )":
                    p_value = 2 * (1 - stats.norm.cdf(abs(Z)))
                    ket_hipotesis = f"H0: p = {p0} vs H1: p â‰  {p0}"
                    tanda = "â‰ "

                elif jenis_uji == "Satu Arah Kanan (>)":
                    p_value = 1 - stats.norm.cdf(Z)
                    ket_hipotesis = f"H0: p â‰¤ {p0} vs H1: p > {p0}"
                    tanda = ">"

                else:
                    p_value = stats.norm.cdf(Z)
                    ket_hipotesis = f"H0: p â‰¥ {p0} vs H1: p < {p0}"
                    tanda = "<"

                # ================= OUTPUT =================
                st.success("âœ… Perhitungan Selesai")

                st.write("#### Hasil Analisis")
                st.markdown(f"**Hipotesis:** {ket_hipotesis}")

                colA, colB, colC = st.columns(3)
                colA.metric("Proporsi Sampel (pÌ‚)", f"{p_hat:.4f}")
                colB.metric("Z-Hitung", f"{Z:.4f}")
                colC.metric("P-Value", f"{p_value:.5f}")

                st.write("#### Kesimpulan")
                if p_value < alpha:
                    st.error("**Keputusan: Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.5f}) < Alpha ({alpha}), "
                        f"maka terdapat cukup bukti bahwa proporsi populasi **{tanda} {p0}**."
                    )
                else:
                    st.warning("**Keputusan: Gagal Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.5f}) â‰¥ Alpha ({alpha}), "
                        f"maka **tidak ada cukup bukti** bahwa proporsi populasi berbeda dari {p0}."
                    )

# UJI PROPORSI 2 SAMPEL
# =========================
def load_proporsi_2_sampel(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    # ================= TAB 1 =================
    with tab1:
        st.subheader("Konsep Uji Proporsi 2 Sampel")
        st.write(r"""
        **Uji proporsi dua sampel** digunakan untuk menguji apakah terdapat 
        **perbedaan proporsi antara dua populasi**.

        Hipotesis yang diuji:
        - $H_0 : p_1 = p_2$
        - $H_1 : p_1 \ne p_2$, atau satu arah
        """)

        st.info(r"""
        **Syarat Penggunaan:**
        1. Dua sampel bersifat **independen**.
        2. Data berupa dikotomi (suksesâ€“gagal).
        3. $n_1p_1, n_1(1-p_1), n_2p_2, n_2(1-p_2) \ge 5$
        """)

    # ================= TAB 2 =================
    with tab2:
        st.subheader("Rumus Statistik Uji Proporsi 2 Sampel")

        st.latex(r"""
        Z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{p(1-p)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
        """)

        st.write("Proporsi Gabungan:")
        st.latex(r"""
        p = \frac{x_1 + x_2}{n_1 + n_2}
        """)

        st.write("Keterangan:")
        st.markdown(r"""
        - $\hat{p}_1 = \frac{x_1}{n_1}$ : proporsi sampel 1  
        - $\hat{p}_2 = \frac{x_2}{n_2}$ : proporsi sampel 2  
        - $x_1, x_2$ : jumlah sukses  
        - $n_1, n_2$ : ukuran sampel  
        """)

    # ================= TAB 3 =================
    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Sebuah perusahaan ingin membandingkan tingkat kelulusan
        antara dua kelas.

        - Kelas A: 40 lulus dari 50 siswa  
        - Kelas B: 32 lulus dari 50 siswa  
        """)

        st.markdown("**Perhitungan:**")
        st.latex(r"\hat{p}_1 = 40/50 = 0.8")
        st.latex(r"\hat{p}_2 = 32/50 = 0.64")
        st.latex(r"p = (40+32)/(100) = 0.72")

        st.latex(r"""
        Z = \frac{0.8 - 0.64}{\sqrt{0.72(0.28)(1/50 + 1/50)}} = 1.75
        """)

        st.write("Karena P-value < 0.05 â†’ **Tolak H0**")

    # ================= TAB 4 =================
    with tab4:
        st.subheader("Kalkulator Uji Proporsi 2 Sampel")

        st.write("### 1. Masukkan Data Sampel")
        col1, col2 = st.columns(2)

        with col1:
            x1 = st.number_input("Jumlah Sukses Sampel 1 (xâ‚)", min_value=0, step=1)
            n1 = st.number_input("Ukuran Sampel 1 (nâ‚)", min_value=1, step=1)

        with col2:
            x2 = st.number_input("Jumlah Sukses Sampel 2 (xâ‚‚)", min_value=0, step=1)
            n2 = st.number_input("Ukuran Sampel 2 (nâ‚‚)", min_value=1, step=1)

        alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])

        jenis_uji = st.selectbox(
            "Jenis Hipotesis Alternatif (H1)",
            ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"]
        )

        st.markdown("---")

        if st.button("Jalankan Uji Proporsi 2 Sampel ğŸš€"):
            if x1 > n1 or x2 > n2:
                st.error("Jumlah sukses tidak boleh lebih besar dari ukuran sampel.")
            else:
                # Hitung proporsi sampel
                p1 = x1 / n1
                p2 = x2 / n2

                # Proporsi gabungan
                p = (x1 + x2) / (n1 + n2)

                # Standard Error
                se = np.sqrt(p * (1 - p) * ((1/n1) + (1/n2)))

                # Statistik Z
                Z = (p1 - p2) / se

                # Hitung p-value
                if jenis_uji == "Dua Arah (â‰ )":
                    p_value = 2 * (1 - stats.norm.cdf(abs(Z)))
                    ket_hipotesis = "H0: p1 = p2 vs H1: p1 â‰  p2"
                    tanda = "â‰ "

                elif jenis_uji == "Satu Arah Kanan (>)":
                    p_value = 1 - stats.norm.cdf(Z)
                    ket_hipotesis = "H0: p1 â‰¤ p2 vs H1: p1 > p2"
                    tanda = ">"

                else:
                    p_value = stats.norm.cdf(Z)
                    ket_hipotesis = "H0: p1 â‰¥ p2 vs H1: p1 < p2"
                    tanda = "<"

                # ================= OUTPUT =================
                st.success("âœ… Perhitungan Selesai")

                st.write("#### Hasil Analisis")
                st.markdown(f"**Hipotesis:** {ket_hipotesis}")

                colA, colB, colC = st.columns(3)
                colA.metric("Proporsi Sampel 1 (pÌ‚â‚)", f"{p1:.4f}")
                colB.metric("Proporsi Sampel 2 (pÌ‚â‚‚)", f"{p2:.4f}")
                colC.metric("Z-Hitung", f"{Z:.4f}")

                st.metric("P-Value", f"{p_value:.6f}")

                st.write("#### Kesimpulan")
                if p_value < alpha:
                    st.error("**Keputusan: Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.6f}) < Alpha ({alpha}), "
                        f"maka terdapat cukup bukti bahwa **pâ‚ {tanda} pâ‚‚**."
                    )
                else:
                    st.warning("**Keputusan: Gagal Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.6f}) â‰¥ Alpha ({alpha}), "
                        f"maka **tidak ada cukup bukti** bahwa proporsi kedua populasi berbeda."
                    )

# UJI RATA - RATA SATU SAMPEL (VARIANS DIKETAHUI) - MENGGUNAKAN Z
def load_z_test_1_sample(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    with tab1:
        st.subheader("Konsep Uji Z 1 Sampel")
        st.write(r"""
        **Uji Z satu sampel** digunakan untuk menguji apakah rata-rata populasi ($\mu$) 
        sama dengan nilai tertentu ($\mu_0$) ketika **varians populasi ($\sigma^2$) diketahui**.
        """)
        st.info(r"""
        **Syarat Penggunaan:**
        1. Data berdistribusi normal (atau $n > 30$).
        2. Varians populasi ($\sigma^2$) atau simpangan baku ($\sigma$) **diketahui**.
        3. Sampel diambil secara acak.
        """)

    with tab2:
        st.subheader("Rumus Statistik Uji Z")
        st.latex(r"""
        Z_{hitung} = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
        """)
        st.write("Keterangan:")
        st.markdown(r"""
        - $\bar{X}$ : Rata-rata sampel
        - $\mu_0$ : Rata-rata hipotesis (nilai acuan)
        - $\sigma$ : Simpangan baku populasi (diketahui)
        - $n$ : Jumlah sampel
        """)

        st.subheader("Kriteria Pengujian (P-value)")
        st.write(r"""
        - **Tolak $H_0$** jika $P\text{-value} < \alpha$
        - **Gagal Tolak $H_0$** jika $P\text{-value} \ge \alpha$
        """)

    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Sebuah pabrik lampu mengklaim umur lampunya rata-rata **800 jam** ($\mu_0$). 
        Diketahui simpangan baku populasi adalah **40 jam** ($\sigma$). 
        Diambil sampel 9 lampu, rata-ratanya **780 jam** ($\bar{X}$). 
        Uji apakah rata-rata umur lampu tersebut berbeda dari klaim pabrik ($\alpha = 5\%$).
        """)
        
        st.markdown("**Perhitungan Manual:**")
        st.latex(r"Z = \frac{780 - 800}{40 / \sqrt{9}} = \frac{-20}{13.33} = -1.5")
        st.write("Karena ini uji dua arah, P-value = $2 \\times P(Z < -1.5) \\approx 0.1336$")
        st.write("Karena $0.1336 > 0.05$, maka **Gagal Tolak H0** (Klaim pabrik diterima).")

    with tab4:
        st.subheader("Kalkulator Uji Z 1 Sampel")
        
        # Tentukan Hipotesis
        st.write("### 1. Tentukan Parameter & Hipotesis")
        
        col1, col2 = st.columns(2)
        with col1:
            mu0 = st.number_input("Rata-rata Hipotesis ($\mu_0$)", value=0.0)
            sigma = st.number_input("Simpangan Baku Populasi ($\sigma$)", value=1.0, min_value=0.0001)
        
        with col2:
            alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10], index=0)
            jenis_uji = st.selectbox("Jenis Hipotesis Alternatif ($H_1$)", 
                                     ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"])

        st.markdown("---")
        
        # Input Data
        st.write("### 2. Input Data Sampel")
        input_metode = st.radio("Pilih metode input:", ["Ketik Manual", "Upload CSV"], horizontal=True)
        
        data_sampel = []
        
        if input_metode == "Ketik Manual":
            input_text = st.text_area("Masukkan data (pisahkan dengan koma)", "10, 12, 11, 14, 13")
            if input_text:
                try:
                    data_sampel = np.array([float(x.strip()) for x in input_text.split(",") if x.strip()])
                except ValueError:
                    st.error("Format salah! Pastikan hanya memasukkan angka dipisahkan koma.")
        else:
            uploaded_file = st.file_uploader("Upload file CSV (pastikan ada kolom header)", type=["csv"])
            if uploaded_file is not None:
                df_upload = pd.read_csv(uploaded_file)
                st.write("Preview Data:", df_upload.head())
                kolom = st.selectbox("Pilih kolom data:", df_upload.columns)
                data_sampel = df_upload[kolom].dropna().values

        # Tombol Hitung 
        if st.button("Jalankan Uji Hipotesis ğŸš€"):
            if len(data_sampel) == 0:
                st.error("Data kosong! Masukkan data terlebih dahulu.")
            else:
                # Hitung Statistik Deskriptif
                n = len(data_sampel)
                x_bar = np.mean(data_sampel)
                
                # Hitung Statistik Uji Z
                se = sigma / np.sqrt(n)  # Standard Error
                z_score = (x_bar - mu0) / se
                
                # Hitung P-value & Tentukan Critical Region
                if jenis_uji == "Dua Arah (â‰ )":
                    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
                    tanda_h1 = "â‰ "
                    ket_hipotesis = f"H0: Î¼ = {mu0} vs H1: Î¼ â‰  {mu0}"
                elif jenis_uji == "Satu Arah Kanan (>)":
                    p_value = 1 - stats.norm.cdf(z_score)
                    tanda_h1 = ">"
                    ket_hipotesis = f"H0: Î¼ â‰¤ {mu0} vs H1: Î¼ > {mu0}"
                else: # Satu Arah Kiri (<)
                    p_value = stats.norm.cdf(z_score)
                    tanda_h1 = "<"
                    ket_hipotesis = f"H0: Î¼ â‰¥ {mu0} vs H1: Î¼ < {mu0}"

                # TAMPILAN OUTPUT
                st.success("âœ… Perhitungan Selesai")
                
                st.write("#### Hasil Analisis")
                st.markdown(f"**Hipotesis:** {ket_hipotesis}")
                
                col_res1, col_res2, col_res3 = st.columns(3)
                col_res1.metric("Rata-rata Sampel (xÌ„)", f"{x_bar:.4f}")
                col_res2.metric("Z-Hitung", f"{z_score:.4f}")
                col_res3.metric("P-Value", f"{p_value:.5f}")
                
                st.write("#### Kesimpulan")
                if p_value < alpha:
                    st.error(f"**Keputusan: Tolak H0**")
                    st.write(f"Karena P-value ({p_value:.5f}) < Alpha ({alpha}), maka ada cukup bukti untuk menyatakan bahwa rata-rata populasi **{tanda_h1} {mu0}**.")
                else:
                    st.warning(f"**Keputusan: Gagal Tolak H0**")
                    st.write(f"Karena P-value ({p_value:.5f}) â‰¥ Alpha ({alpha}), maka **tidak ada cukup bukti** untuk menyatakan rata-rata populasi berbeda dari {mu0}.")


# UJI RATA - RATA SATU SAMPEL (Varians tidak diketahui) menggunakan t
def load_t_test_1_sample(title):
    st.header(title)
    
    # Membuat Tabs
    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    with tab1:
        st.subheader("Konsep Uji t 1 Sampel")
        st.write(r"""
        **Uji t satu sampel** digunakan ketika kita ingin menguji rata-rata populasi ($\mu$) 
        tetapi **varians populasi ($\sigma^2$) tidak diketahui**. 
        Sebagai gantinya, kita menggunakan varians sampel ($s^2$).
        """)
        st.info("""
        **Syarat:**
        1. Varians populasi tidak diketahui.
        2. Sampel diambil secara acak.
        3. Data berdistribusi normal.
        """)

    with tab2:
        st.subheader("Rumus Statistik Uji t")
        st.latex(r"t_{hitung} = \frac{\bar{X} - \mu_0}{s / \sqrt{n}}")
        
        st.write("Derajat Bebas ($df$):")
        st.latex(r"df = n - 1")
        
        st.write("Keterangan:")
        st.markdown(r"""
        - $\bar{X}$: Rata-rata sampel
        - $\mu_0$: Rata-rata hipotesis
        - $s$: Standar deviasi sampel
        - $n$: Jumlah sampel
        """)

    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Seorang peneliti menduga rata-rata berat badan mahasiswa adalah **60 kg**. 
        Diambil sampel 5 orang: **62, 65, 58, 61, 64**. 
        Uji apakah rata-rata berat badan berbeda dari 60 kg ($\alpha=0.05$).
        """)
        st.markdown("**Langkah:**")
        st.write("1. Rata-rata ($X$) = 62")
        st.write("2. Standar Deviasi ($s$) = 2.738")
        st.write("3. $t_{hitung} = (62 - 60) / (2.738 / \sqrt{5}) = 1.633$")

    with tab4:
        st.subheader("Kalkulator Uji t (Sesuai Flowchart)")
        
        st.write("#### 1. Tentukan Hipotesis")
        col1, col2 = st.columns(2)
        with col1:
            mu0 = st.number_input(r"Nilai Acuan ($\mu_0$)", value=0.0)
        with col2:
            alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])
            jenis_uji = st.selectbox("Jenis Uji ($H_1$)", ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"])

        st.write("#### 2. Input Data Sampel")
        input_text = st.text_area("Masukkan data (pisahkan dengan koma)", "62, 65, 58, 61, 64")
        
        if st.button("Mulai Perhitungan ğŸš€"):
            try:
                # Proses Data
                data = np.array([float(x.strip()) for x in input_text.split(",") if x.strip()])
                n = len(data)
                
                # 1. Hitung Rata-rata Sampel
                x_bar = np.mean(data)
                
                # 2. Hitung Standar Deviasi Sampel (s)
                s = np.std(data, ddof=1) # ddof=1 untuk sampel
                
                # 3. Hitung Derajat Bebas (df)
                df = n - 1
                
                # 4. Hitung Statistik Uji t
                se = s / np.sqrt(n)
                t_hitung = (x_bar - mu0) / se
                
                # 5. Hitung P-value / Titik Kritis
                if jenis_uji == "Dua Arah (â‰ )":
                    p_value = 2 * (1 - stats.t.cdf(abs(t_hitung), df))
                    tanda = "â‰ "
                elif jenis_uji == "Satu Arah Kanan (>)":
                    p_value = 1 - stats.t.cdf(t_hitung, df)
                    tanda = ">"
                else:
                    p_value = stats.t.cdf(t_hitung, df)
                    tanda = "<"
                
                # Tampilkan Output Bertahap
                st.markdown("---")
                st.success("âœ… Hasil Perhitungan")
                
                col_A, col_B, col_C = st.columns(3)
                col_A.metric(r"Rata-rata ($\bar{X}$)", f"{x_bar:.4f}")
                col_B.metric(r"Std. Deviasi ($s$)", f"{s:.4f}")
                col_C.metric(r"Derajat Bebas ($df$)", f"{df}")
                
                st.metric(r"t-hitung ($t$)", f"{t_hitung:.4f}")
                st.metric("P-value", f"{p_value:.5f}")
                
                # Kesimpulan
                st.write("#### Kesimpulan Hipotesis")
                if p_value < alpha:
                    st.error(f"**Tolak H0**")
                    st.write(f"Karena P-value ({p_value:.5f}) < Alpha ({alpha}), maka rata-rata populasi **{tanda} {mu0}**.")
                else:
                    st.warning(f"**Gagal Tolak H0**")
                    st.write(f"Karena P-value ({p_value:.5f}) â‰¥ Alpha ({alpha}), maka tidak cukup bukti rata-rata berbeda.")
                    
            except ValueError:
                st.error("Pastikan data yang dimasukkan hanya angka dipisahkan koma.")


# F-test
def load_f_test(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["Penjelasan", "Rumus", "Contoh", "Output Perhitungan"]
    )

    with tab1:
        st.subheader("ğŸ“˜ Penjelasan")
        st.write("""
        *Uji F (F-test)* digunakan untuk menguji apakah *dua varians populasi sama atau berbeda*.
        
        Biasanya dilakukan sebelum uji t dua sampel,
        untuk menentukan apakah memakai *Pooled t-test* atau *Welch t-test*.
        """)

        st.info("""
        *Asumsi F-test:*
        1. Data dari dua sampel berdistribusi normal.
        2. Dua sampel independen.
        """)

    with tab2:
        st.subheader("ğŸ“ Rumus Statistik Uji")
        
        st.latex(r"""
        F = \frac{S_1^2}{S_2^2}
        """)

        st.write("Keterangan:")
        st.markdown("""
        - $S_1^2$ : Varians sampel 1  
        - $S_2^2$ : Varians sampel 2  
        - df1 = $n_1 - 1$  
        - df2 = $n_2 - 1$  
        """)

        st.warning("Catatan: varians yang lebih besar biasanya ditempatkan sebagai pembilang, agar F â‰¥ 1.")

    with tab3:
        st.subheader("ğŸ§® Contoh Perhitungan")

        st.write("""
        Misalkan data panjang baja dari dua pabrik sebagai berikut:
        """)

        data1 = [7, 9, 12, 10, 11]
        data2 = [8, 8, 9, 7, 6]

        df = pd.DataFrame({
            "Pabrik A": data1,
            "Pabrik B": data2
        })

        st.dataframe(df)

        var1 = np.var(data1, ddof=1)
        var2 = np.var(data2, ddof=1)

        st.write(f"Varians Pabrik A = {var1:.3f}")
        st.write(f"Varians Pabrik B = {var2:.3f}")
        st.write(f"Rasio F = {var1/var2:.3f}")

    with tab4:
        st.subheader("Kalkulator F-Test")

        col1, col2 = st.columns(2)

        with col1:
            input1 = st.text_area("Data Sampel 1")

        with col2:
            input2 = st.text_area("Data Sampel 2")

        alpha = st.selectbox("Alpha", [0.05, 0.01, 0.10])

        if st.button("Hitung F-test"):
            try:
                x1 = np.array([float(x) for x in input1.split(",")])
                x2 = np.array([float(x) for x in input2.split(",")])

                var1 = np.var(x1, ddof=1)
                var2 = np.var(x2, ddof=1)

                F = var1 / var2 if var1 >= var2 else var2 / var1

                df1 = len(x1) - 1
                df2 = len(x2) - 1

                p_value = 1 - stats.f.cdf(F, df1, df2)

                colA, colB, colC = st.columns(3)
                colA.metric("F-hitung", f"{F:.4f}")
                colB.metric("df1, df2", f"{df1}, {df2}")
                colC.metric("p-value", f"{p_value:.5f}")

                if p_value < alpha:
                    st.success("Tolak H0 â€” Varians berbeda.")
                else:
                    st.warning("Gagal Tolak H0 â€” Varians sama.")

            except:
                st.error("Format input tidak valid.")


# UJI RATA-RATA 2 SAMPEL INDEPENDEN (Z-TEST)
# =========================
def load_z_2_sample(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    # ================= TAB 1 =================
    with tab1:
        st.subheader("Konsep Uji Z Dua Sampel Independen")
        st.write(r"""
        **Uji Z dua sampel independen** digunakan untuk menguji apakah terdapat
        **perbedaan rata-rata antara dua populasi** ketika **varians populasi diketahui**.

        Hipotesis:
        - $H_0 : \mu_1 = \mu_2$
        - $H_1 : \mu_1 \ne \mu_2$ atau satu arah
        """)

        st.info(r"""
        **Syarat Penggunaan:**
        1. Dua sampel **independen**.
        2. Data berdistribusi normal atau $n_1, n_2 \ge 30$.
        3. Varians populasi ($\sigma_1^2, \sigma_2^2$) **diketahui**.
        """)

    # ================= TAB 2 =================
    with tab2:
        st.subheader("Rumus Statistik Uji Z Dua Sampel")

        st.latex(r"""
        Z = \frac{\bar{X}_1 - \bar{X}_2}
        {\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}
        """)

        st.write("Keterangan:")
        st.markdown(r"""
        - $\bar{X}_1, \bar{X}_2$ : rata-rata sampel  
        - $\sigma_1, \sigma_2$ : standar deviasi populasi  
        - $n_1, n_2$ : ukuran sampel  
        """)

        st.subheader("Kriteria Keputusan")
        st.write(r"""
        - **Tolak $H_0$** jika $P\text{-value} < \alpha$
        - **Gagal Tolak $H_0$** jika $P\text{-value} \ge \alpha$
        """)

    # ================= TAB 3 =================
    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Sebuah penelitian membandingkan rata-rata nilai matematika
        dua kelas dengan data berikut:

        - $\bar{X}_1 = 78$, $n_1 = 40$, $\sigma_1 = 8$  
        - $\bar{X}_2 = 74$, $n_2 = 35$, $\sigma_2 = 7$
        """)

        st.latex(r"""
        Z = \frac{78 - 74}{\sqrt{8^2/40 + 7^2/35}} = 2.12
        """)

        st.write("Karena P-value < 0.05 â†’ **Tolak H0**")

    # ================= TAB 4 =================
    with tab4:
        st.subheader("Kalkulator Uji Z Dua Sampel Independen")

        st.write("### 1. Input Parameter Sampel")

        col1, col2 = st.columns(2)
        with col1:
            xbar1 = st.number_input("Rata-rata Sampel 1 (XÌ„â‚)", value=0.0)
            sigma1 = st.number_input("Simpangan Baku Populasi 1 (Ïƒâ‚)", value=1.0, min_value=0.0001)
            n1 = st.number_input("Ukuran Sampel 1 (nâ‚)", min_value=1, step=1)

        with col2:
            xbar2 = st.number_input("Rata-rata Sampel 2 (XÌ„â‚‚)", value=0.0)
            sigma2 = st.number_input("Simpangan Baku Populasi 2 (Ïƒâ‚‚)", value=1.0, min_value=0.0001)
            n2 = st.number_input("Ukuran Sampel 2 (nâ‚‚)", min_value=1, step=1)

        alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])

        jenis_uji = st.selectbox(
            "Jenis Hipotesis Alternatif (H1)",
            ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"]
        )

        st.markdown("---")

        if st.button("Jalankan Uji Z Dua Sampel ğŸš€"):
            # Hitung Standard Error
            se = np.sqrt((sigma1**2 / n1) + (sigma2**2 / n2))

            # Hitung Z
            Z = (xbar1 - xbar2) / se

            # Hitung p-value
            if jenis_uji == "Dua Arah (â‰ )":
                p_value = 2 * (1 - stats.norm.cdf(abs(Z)))
                ket_hipotesis = "H0: Î¼1 = Î¼2 vs H1: Î¼1 â‰  Î¼2"
                tanda = "â‰ "

            elif jenis_uji == "Satu Arah Kanan (>)":
                p_value = 1 - stats.norm.cdf(Z)
                ket_hipotesis = "H0: Î¼1 â‰¤ Î¼2 vs H1: Î¼1 > Î¼2"
                tanda = ">"

            else:
                p_value = stats.norm.cdf(Z)
                ket_hipotesis = "H0: Î¼1 â‰¥ Î¼2 vs H1: Î¼1 < Î¼2"
                tanda = "<"

            # ================= OUTPUT =================
            st.success("âœ… Perhitungan Selesai")

            st.write("#### Hasil Analisis")
            st.markdown(f"**Hipotesis:** {ket_hipotesis}")

            colA, colB, colC = st.columns(3)
            colA.metric("Z-Hitung", f"{Z:.4f}")
            colB.metric("Alpha", f"{alpha}")
            colC.metric("P-Value", f"{p_value:.6f}")

            st.write("#### Kesimpulan")
            if p_value < alpha:
                st.error("**Keputusan: Tolak H0**")
                st.write(
                    f"Karena P-value ({p_value:.6f}) < Alpha ({alpha}), "
                    f"maka terdapat cukup bukti bahwa **Î¼â‚ {tanda} Î¼â‚‚**."
                )
            else:
                st.warning("**Keputusan: Gagal Tolak H0**")
                st.write(
                    f"Karena P-value ({p_value:.6f}) â‰¥ Alpha ({alpha}), "
                    f"maka **tidak ada cukup bukti** bahwa rata-rata kedua populasi berbeda."
                )


# UJI RATA-RATA 2 SAMPEL INDEPENDEN (POOLED T-TEST)
# =========================
def load_pooled_t_test(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    # ================= TAB 1 =================
    with tab1:
        st.subheader("Konsep Pooled t-Test")
        st.write(r"""
        **Pooled t-test** digunakan untuk menguji apakah terdapat
        **perbedaan rata-rata dua populasi independen**
        dengan **asumsi varians kedua populasi sama**.

        Hipotesis:
        - $H_0 : \mu_1 = \mu_2$
        - $H_1 : \mu_1 \ne \mu_2$ atau satu arah
        """)

        st.info(r"""
        **Syarat Penggunaan:**
        1. Dua sampel **independen**.
        2. Data berdistribusi normal.
        3. Varians kedua populasi **diasumsikan sama**.
        4. Skala data numerik.
        """)

    # ================= TAB 2 =================
    with tab2:
        st.subheader("Rumus Statistik Pooled t-Test")

        st.latex(r"""
        t = \frac{\bar{X}_1 - \bar{X}_2}
        {S_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
        """)

        st.write("Rumus Varians Gabungan (Pooled Variance):")
        st.latex(r"""
        S_p^2 = \frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}
        """)

        st.write("Derajat Bebas:")
        st.latex(r"""
        df = n_1 + n_2 - 2
        """)

        st.write("Keterangan:")
        st.markdown(r"""
        - $\bar{X}_1, \bar{X}_2$ : rata-rata sampel  
        - $s_1, s_2$ : standar deviasi sampel  
        - $n_1, n_2$ : ukuran sampel  
        """)

    # ================= TAB 3 =================
    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Dua kelas diuji nilai matematikanya:

        - Kelas A: 70, 75, 78, 74, 76  
        - Kelas B: 65, 68, 72, 70, 69  
        """)

        A = np.array([70, 75, 78, 74, 76])
        B = np.array([65, 68, 72, 70, 69])

        x1, x2 = np.mean(A), np.mean(B)
        s1, s2 = np.var(A, ddof=1), np.var(B, ddof=1)
        n1, n2 = len(A), len(B)

        sp2 = ((n1-1)*s1 + (n2-1)*s2) / (n1+n2-2)
        t_hit = (x1 - x2) / (np.sqrt(sp2) * np.sqrt(1/n1 + 1/n2))

        st.write(f"Rata-rata Kelas A = {x1:.2f}")
        st.write(f"Rata-rata Kelas B = {x2:.2f}")
        st.write(f"t-hitung = {t_hit:.4f}")

        st.write("Jika P-value < 0.05 â†’ **Tolak H0**")

    # ================= TAB 4 =================
    with tab4:
        st.subheader("Kalkulator Pooled t-Test")

        st.write("### 1. Input Data Sampel")

        col1, col2 = st.columns(2)
        with col1:
            data1_text = st.text_area("Data Sampel 1 (pisahkan dengan koma)",
                                      "70,75,78,74,76")
        with col2:
            data2_text = st.text_area("Data Sampel 2 (pisahkan dengan koma)",
                                      "65,68,72,70,69")

        alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])
        jenis_uji = st.selectbox(
            "Jenis Hipotesis Alternatif (H1)",
            ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"]
        )

        st.markdown("---")

        if st.button("Jalankan Pooled t-Test ğŸš€"):
            try:
                X1 = np.array([float(i.strip()) for i in data1_text.split(",") if i.strip()])
                X2 = np.array([float(i.strip()) for i in data2_text.split(",") if i.strip()])

                n1, n2 = len(X1), len(X2)
                xbar1, xbar2 = np.mean(X1), np.mean(X2)
                s1, s2 = np.var(X1, ddof=1), np.var(X2, ddof=1)

                # Varians Gabungan
                sp2 = ((n1-1)*s1 + (n2-1)*s2) / (n1+n2-2)
                sp = np.sqrt(sp2)

                # Statistik t
                t_hit = (xbar1 - xbar2) / (sp * np.sqrt(1/n1 + 1/n2))

                # Derajat Bebas
                df = n1 + n2 - 2

                # P-value
                if jenis_uji == "Dua Arah (â‰ )":
                    p_value = 2 * (1 - stats.t.cdf(abs(t_hit), df))
                    ket_hipotesis = "H0: Î¼1 = Î¼2 vs H1: Î¼1 â‰  Î¼2"
                    tanda = "â‰ "
                elif jenis_uji == "Satu Arah Kanan (>)":
                    p_value = 1 - stats.t.cdf(t_hit, df)
                    ket_hipotesis = "H0: Î¼1 â‰¤ Î¼2 vs H1: Î¼1 > Î¼2"
                    tanda = ">"
                else:
                    p_value = stats.t.cdf(t_hit, df)
                    ket_hipotesis = "H0: Î¼1 â‰¥ Î¼2 vs H1: Î¼1 < Î¼2"
                    tanda = "<"

                # ================= OUTPUT =================
                st.success("âœ… Perhitungan Selesai")

                st.write("#### Hasil Analisis")
                st.markdown(f"**Hipotesis:** {ket_hipotesis}")

                colA, colB, colC = st.columns(3)
                colA.metric("Rata-rata Sampel 1", f"{xbar1:.4f}")
                colB.metric("Rata-rata Sampel 2", f"{xbar2:.4f}")
                colC.metric("t-Hitung", f"{t_hit:.4f}")

                st.metric("P-Value", f"{p_value:.6f}")
                st.metric("Derajat Bebas (df)", f"{df}")

                st.write("#### Kesimpulan")
                if p_value < alpha:
                    st.error("**Keputusan: Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.6f}) < Alpha ({alpha}), "
                        f"maka terdapat cukup bukti bahwa **Î¼â‚ {tanda} Î¼â‚‚**."
                    )
                else:
                    st.warning("**Keputusan: Gagal Tolak H0**")
                    st.write(
                        f"Karena P-value ({p_value:.6f}) â‰¥ Alpha ({alpha}), "
                        f"maka **tidak ada cukup bukti** bahwa rata-rata kedua populasi berbeda."
                    )

            except:
                st.error("Format data salah. Pastikan hanya angka dipisahkan koma.")

                
#UJI WELSCH T TEST ==================================================

def load_template_welch_t_test(title):
    # Judul Utama
    st.header(title)
    
    # Tabs
    tab1, tab2, tab3, tab4 = st.tabs(
        ["Penjelasan", "Rumus", "Contoh", "Output Perhitungan"]
    )

    # --- TAB 1: PENJELASAN ---
    with tab1:
        st.subheader("ğŸ“˜ Penjelasan (Welch's T-Test)")
        st.info("""
        digunakan untuk membandingkan rata-rata dari dua kelompok independen ketika varians populasi kedua kelompok diasumsikan tidak sama
        
        **Kapan Menggunakan Welch's T-Test (Syarat Kunci):**
        1.  **Dua Kelompok Independen:** Data berasal dari dua kelompok yang tidak berpasangan.
        3.  **Distribusi Normal:** Data terdistribusi normal (atau ukuran sampel cukup besar).
        4.  **Varians Tidak Sama** Ini adalah kondisi pembeda
        """)

    # --- TAB 2: RUMUS ---
    with tab2:
        st.subheader("ğŸ“ Rumus Statistik Uji")
        
        # Rumus Statistik t
        st.markdown("**1. Statistik t (t-statistic):**")
        st.latex(r"""
            t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
        """)
        
        # Keterangan Rumus
        st.markdown(r"""
        * $\bar{X}_1, \bar{X}_2$: Rata-rata sampel kelompok 1 dan 2.
        * $s_1^2, s_2^2$: Varians sampel kelompok 1 dan 2.
        * $n_1, n_2$: Ukuran sampel kelompok 1 dan 2.
        """)
        
        # Rumus Derajat Kebebasan (df)
        st.markdown("**2. Derajat Kebebasan (df) :**")
        st.latex(r"""
            \nu = \text{df} = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(s_1^2/n_1\right)^2}{n_1 - 1} + \frac{\left(s_2^2/n_2\right)^2}{n_2 - 1}}
        """)
        

    # --- TAB 3: CONTOH PERHITUNGAN ---
    with tab3:
        st.subheader("ğŸ§® Contoh Perhitungan")
        st.markdown("""
**Contoh Kasus â€“ Uji Welch t-test**

Seorang peneliti ingin mengetahui apakah terdapat perbedaan 
rata-rata nilai ujian Matematika antara **Kelas A** dan **Kelas B**.

- **Kelas A:** 78, 85, 80, 90, 88, 76, 82, 84  
- **Kelas B:** 70, 72, 68, 75, 73, 77  
        """)

        # tampilkan data contoh
        kelas_A = np.array([78, 85, 80, 90, 88, 76, 82, 84])
        kelas_B = np.array([70, 72, 68, 75, 73, 77])
        st.write("Data Kelas A:", kelas_A)
        st.write("Data Kelas B:", kelas_B)

        # hitung welch t-test di tab contoh juga
        t_stat, p_value = stats.ttest_ind(kelas_A, kelas_B, equal_var=False)
        # df manual
        n1, n2 = len(kelas_A), len(kelas_B)
        s1, s2 = np.var(kelas_A, ddof=1), np.var(kelas_B, ddof=1)
        df = (s1/n1 + s2/n2)**2 / ((s1**2)/((n1**2)*(n1-1)) + (s2**2)/((n2**2)*(n2-1)))

        st.write("t-statistic:", round(t_stat, 4))
        st.write("p-value:", round(p_value, 6))
        st.write("Derajat kebebasan (Welch df):", round(df, 4))

        if p_value < 0.05:
            st.success("Hâ‚€ ditolak â€” ada perbedaan signifikan (Î±=0.05).")
        else:
            st.info("Gagal menolak Hâ‚€ â€” tidak ada bukti perbedaan signifikan.")

    

    # --- TAB 4: OUTPUT PERHITUNGAN OTOMATIS ---
    with tab4:
        st.subheader("ğŸŸ¢ Perhitungan Otomatis (Menggunakan Data Input)")
        st.write("Silakan masukkan data Anda untuk menghitung Welch's T-Test secara otomatis.")

        # Input data (Placeholder for user input)
        st.markdown("#### Taraf Signifikansi dan Data Sampel")
        
        # Penambahan kustomisasi Taraf Signifikansi (Alpha) - Label diubah
        alpha = st.number_input(
            "Taraf Signifikansi (Î±):", # Menggunakan simbol Unicode 'Î±'
            min_value=0.001, 
            max_value=0.5, 
            value=0.05, 
            step=0.005,
            format="%.3f",
            help="Taraf maksimum risiko kesalahan Tipe I yang diterima (biasanya 0.05)."
        )

       
        data_a_str = st.text_input("Data Program A (Kelompok 1):", "84, 85, 86, 83, 85, 87, 84, 85, 86, 85, 84, 85, 86, 83, 85, 87, 84, 85, 86, 85") 
        
        data_b_str = st.text_input("Data Program B (Kelompok 2):", "70, 95, 65, 80, 90, 75, 50, 99, 81, 78, 60, 100, 72, 88, 77") 

        try:
            # Konversi string -> array
            data_A = np.array([float(x.strip()) for x in data_a_str.split(',') if x.strip()])
            data_B = np.array([float(x.strip()) for x in data_b_str.split(',') if x.strip()])

            if len(data_A) > 1 and len(data_B) > 1:
                # Welch T-test
                t_stat_auto, p_value_auto = stats.ttest_ind(
                    a=data_A, 
                    b=data_B, 
                    equal_var=False 
                )
                
                # Statistik Deskriptif
                df_deskriptif = pd.DataFrame({
                    'Kelompok': ['A', 'B'],
                    'Rata-rata (XÌ„)': [np.mean(data_A), np.mean(data_B)],
                    'Varians (sÂ²)': [np.var(data_A, ddof=1), np.var(data_B, ddof=1)],
                    'Ukuran Sampel (n)': [len(data_A), len(data_B)]
                })
                st.dataframe(df_deskriptif, hide_index=True)


                st.markdown("#### Hasil Welch's T-Test")
                st.code(f"""
                Statistik t Welch: {t_stat_auto:.4f}
                Nilai p (Dua Arah): {p_value_auto:.4f}
                Derajat Kebebasan (df): Diperkirakan oleh SciPy
                """, language='text')

                
                # Menggunakan simbol LaTeX di sini aman karena berada dalam st.markdown
                if p_value_auto < alpha:
                    st.error(f"P-value ({p_value_auto:.4f}) < Taraf Signifikansi ($\\alpha$ = {alpha:.3f}). Tolak H0.")
                    st.markdown("##### **Kesimpulan:** Ada perbedaan rata-rata yang signifikan secara statistik antara Kelompok A dan Kelompok B.")
                else:
                    st.success(f"P-value ({p_value_auto:.4f}) â‰¥ Taraf Signifikansi ($\\alpha$ = {alpha:.3f}). Gagal Tolak H0.")
                    st.markdown("##### **Kesimpulan:** Tidak ada bukti yang cukup untuk menyatakan perbedaan rata-rata yang signifikan antara Kelompok A dan Kelompok B.")

            else:
                st.warning("Masukkan minimal dua nilai di setiap kelompok untuk melakukan perhitungan.")
        
        except ValueError:
            st.error("Pastikan data yang dimasukkan hanya berupa angka dan dipisahkan oleh koma.")


# UJI RATA-RATA 2 SAMPEL DEPENDEN (PAIRED T-TEST)
# =========================
def load_paired_t_test(title):
    st.header(title)

    tab1, tab2, tab3, tab4 = st.tabs(
        ["ğŸ“˜ Penjelasan", "ğŸ“ Rumus", "ğŸ§® Contoh", "ğŸŸ¢ Perhitungan"]
    )

    # ================= TAB 1 =================
    with tab1:
        st.subheader("Konsep Paired t-Test")
        st.write(r"""
        **Paired t-test (uji t berpasangan)** digunakan untuk menguji apakah
        terdapat **perbedaan rata-rata sebelum dan sesudah perlakuan**
        pada **subjek yang sama**.

        Contoh penggunaan:
        - Nilai sebelum dan sesudah pelatihan
        - Berat badan sebelum dan sesudah diet

        Hipotesis:
        - $H_0 : \mu_d = 0$
        - $H_1 : \mu_d \ne 0$ atau satu arah
        """)

        st.info(r"""
        **Syarat Penggunaan:**
        1. Data berpasangan (subjek sama).
        2. Selisih data berdistribusi normal.
        3. Skala data numerik.
        """)

    # ================= TAB 2 =================
    with tab2:
        st.subheader("Rumus Statistik Paired t-Test")

        st.latex(r"""
        t = \frac{\bar{d} - \mu_0}{s_d / \sqrt{n}}
        """)

        st.write("Keterangan:")
        st.markdown(r"""
        - $d_i = X_{1i} - X_{2i}$ : selisih berpasangan  
        - $\bar{d}$ : rata-rata selisih  
        - $s_d$ : standar deviasi selisih  
        - $n$ : jumlah pasangan  
        - $\mu_0$ : selisih hipotesis (biasanya 0)
        """)

        st.write("Derajat Bebas:")
        st.latex(r"df = n - 1")

    # ================= TAB 3 =================
    with tab3:
        st.subheader("Contoh Kasus")
        st.write(r"""
        Nilai mahasiswa sebelum dan sesudah bimbingan:

        | Sebelum | Sesudah |
        |----------|----------|
        | 70 | 75 |
        | 65 | 70 |
        | 80 | 85 |
        | 75 | 78 |
        | 72 | 76 |
        """)

        sebelum = np.array([70, 65, 80, 75, 72])
        sesudah = np.array([75, 70, 85, 78, 76])
        d = sebelum - sesudah

        dbar = np.mean(d)
        sd = np.std(d, ddof=1)
        n = len(d)

        t_hit = dbar / (sd / np.sqrt(n))

        st.write(f"Rata-rata selisih = {dbar:.2f}")
        st.write(f"t-hitung = {t_hit:.4f}")

        st.write("Jika P-value < 0.05 â†’ **Tolak H0**")

    # ================= TAB 4 =================
    with tab4:
        st.subheader("Kalkulator Paired t-Test")

        st.write("### 1. Input Data Berpasangan")

        col1, col2 = st.columns(2)
        with col1:
            data1_text = st.text_area("Data Sebelum",
                                      "70,65,80,75,72")
        with col2:
            data2_text = st.text_area("Data Sesudah",
                                      "75,70,85,78,76")

        mu0 = st.number_input("Nilai Hipotesis Selisih (Î¼â‚€)", value=0.0)

        alpha = st.selectbox(r"Taraf Signifikansi ($\alpha$)", [0.05, 0.01, 0.10])

        jenis_uji = st.selectbox(
            "Jenis Hipotesis Alternatif (H1)",
            ["Dua Arah (â‰ )", "Satu Arah Kanan (>)", "Satu Arah Kiri (<)"]
        )

        st.markdown("---")

        if st.button("Jalankan Paired t-Test ğŸš€"):
            try:
                X1 = np.array([float(i.strip()) for i in data1_text.split(",") if i.strip()])
                X2 = np.array([float(i.strip()) for i in data2_text.split(",") if i.strip()])

                if len(X1) != len(X2):
                    st.error("Jumlah data sebelum dan sesudah harus sama!")
                else:
                    D = X1 - X2
                    n = len(D)
                    dbar = np.mean(D)
                    sd = np.std(D, ddof=1)

                    t_hit = (dbar - mu0) / (sd / np.sqrt(n))

                    df = n - 1

                    # Hitung P-value
                    if jenis_uji == "Dua Arah (â‰ )":
                        p_value = 2 * (1 - stats.t.cdf(abs(t_hit), df))
                        ket_hipotesis = "H0: Î¼d = 0 vs H1: Î¼d â‰  0"
                        tanda = "â‰ "
                    elif jenis_uji == "Satu Arah Kanan (>)":
                        p_value = 1 - stats.t.cdf(t_hit, df)
                        ket_hipotesis = "H0: Î¼d â‰¤ 0 vs H1: Î¼d > 0"
                        tanda = ">"
                    else:
                        p_value = stats.t.cdf(t_hit, df)
                        ket_hipotesis = "H0: Î¼d â‰¥ 0 vs H1: Î¼d < 0"
                        tanda = "<"

                    # ================= OUTPUT =================
                    st.success("âœ… Perhitungan Selesai")

                    st.write("#### Hasil Analisis")
                    st.markdown(f"**Hipotesis:** {ket_hipotesis}")

                    colA, colB, colC = st.columns(3)
                    colA.metric("Rata-rata Selisih (dÌ„)", f"{dbar:.4f}")
                    colB.metric("Std. Deviasi Selisih (sd)", f"{sd:.4f}")
                    colC.metric("t-Hitung", f"{t_hit:.4f}")

                    st.metric("P-Value", f"{p_value:.6f}")
                    st.metric("Derajat Bebas (df)", f"{df}")

                    st.write("#### Kesimpulan")
                    if p_value < alpha:
                        st.error("**Keputusan: Tolak H0**")
                        st.write(
                            f"Karena P-value ({p_value:.6f}) < Alpha ({alpha}), "
                            f"maka terdapat cukup bukti bahwa **Î¼d {tanda} 0**."
                        )
                    else:
                        st.warning("**Keputusan: Gagal Tolak H0**")
                        st.write(
                            f"Karena P-value ({p_value:.6f}) â‰¥ Alpha ({alpha}), "
                            f"maka **tidak ada cukup bukti** bahwa terjadi perbedaan yang signifikan."
                        )

            except:
                st.error("Format data salah. Pastikan hanya angka dipisahkan koma.")


if main_menu == "ğŸ“š Pilih Uji Statistik":

    if menu == "Uji Proporsi 1 Sampel":
        load_proporsi_1_sampel("Uji Proporsi 1 Sampel")

    elif menu == "Uji Proporsi 2 Sampel":
        load_proporsi_2_sampel("Uji Proporsi 2 Sampel")

    elif menu == "Uji Rata-rata 1 Sampel (Z)":
        load_z_test_1_sample("Uji Rata-rata 1 Sampel (Z-Test)")

    elif menu == "Uji Rata-rata 1 Sampel (t)":
        load_t_test_1_sample("Uji Rata-rata 1 Sampel (t-Test)")

    elif menu == "Uji Kesamaan Varians (F-test)":
        load_f_test("Uji Kesamaan Varians (F-test)")

    elif menu == "Uji Rata-rata 2 Sampel Independen (Z-test)":
        load_z_2_sample("Uji Rata-rata 2 Sampel Independen (Z-test)")

    elif menu == "Uji Rata-rata 2 Sampel Independen (Pooled t-test)":
        load_pooled_t_test("Uji Rata-rata 2 Sampel Independen (Pooled t-test)")

    elif menu == "Uji Rata-rata 2 Sampel Independen (Welch t-test)":
        load_template_welch_t_test("Uji Rata-rata 2 Sampel Independen (Welch t-test)")

    elif menu == "Uji Rata-rata 2 Sampel Dependen (Paired t-test)":
        load_paired_t_test("Uji Rata-rata 2 Sampel Dependen (Paired t-test)")
```
